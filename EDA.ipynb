# Exploratory Data Analysis (EDA)

## Business Questions

Before analyzing the dataset, the following key questions are considered:

1. Which product category generates the highest sales?
2. Which region contributes the most profit?
3. How does discount impact profit?
4. Which sub-category experiences the highest losses?
5. What are the sales trends over time?
6. Which customer segment generates the most revenue?

These questions guide the analysis and help uncover meaningful insights from the dataset.

```python
import pandas as pd

df = pd.read_csv("Sample - Superstore.csv", encoding='latin1')

df.head()
```

```python
df.shape
df.columns
df.info()

df.describe()

df.isnull().sum()
```

## Data Structure Overview

* The dataset contains 9994 rows and 21 columns.
* There are no missing values in the dataset.
* The dataset consists of both numerical and categorical variables.

  * Numerical features include Sales, Profit, Quantity, and Discount.
  * Categorical features include Category, Region, Segment, and Sub-Category.


```python
import matplotlib.pyplot as plt
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

Observation:
Technology category has the highest sales.

```python
import seaborn as sns

sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

Observation:
As discount increases, profit decreases.

```python
numeric_df = df.select_dtypes(include=['number'])

sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```

## Correlation Analysis

The correlation heatmap shows relationships between numerical variables.

* Profit is negatively correlated with Discount.
* Sales and Profit show a weak positive relationship.
* Discount has a significant impact on Profit.

This indicates that higher discounts may reduce overall profitability.


```python
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.show()
```

## Hypothesis Testing

Hypothesis:
Higher discounts lead to lower profit.

Result:
The scatter plot shows a negative relationship between discount and profit.

Conclusion:
The hypothesis is valid.
